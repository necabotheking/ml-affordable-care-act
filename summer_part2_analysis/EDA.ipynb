```python
import pandas as pd
import matplotlib.pyplot as plt
import re
from nltk.corpus import stopwords
import nltk
from nltk.tokenize import wordpunct_tokenize
nltk.download('stopwords')
import matplotlib.pyplot as plt
```

```python
tweets = pd.read_csv('obamacare_19_23.csv')
```

```python
tweets.loc[:, 'Text_Split'] = tweets.loc[:, 'Text'].apply(lambda x: x.split())
```

#### Most Retweeted Tweets

```python

pd.set_option('display.max_colwidth', None)

text_counts = tweets['Text'].value_counts()

# retweeted tweets 
repeated_text = text_counts[text_counts > 1]

for text, count in repeated_text.items():
    print("Count:", count)
    print("Text:", text)
    print("")
```

#### Tweets that contain 'obamacare'
* In order to be included, the tweet just must simply contain the string 'obamacare' in the text.

```python
obamacare = tweets[tweets['Text'].str.contains('obamacare', case=False)]
print(obamacare)
```

```python
stop = set(stopwords.words('english'))
stop.update(['.', ',', '"', "'", '?', '!', ':', ';', '(', ')', '[', ']', '{', '}', '@', '#', 'rt', 'amp', 'http', 'https', '/', '://', '_', 'co', 'RT'])
```

```python
obamacare = tweets[tweets['Text'].str.contains('obamacare', case=False)]
print(obamacare)
```

```python
series_tweets = obamacare['Text']
tweet_str = series_tweets.str.cat(sep = ' ')
list_of_words = [i.lower() for i in wordpunct_tokenize(tweet_str) if i.lower() not in stop and i.isalpha()]
wordfreqdist = nltk.FreqDist(list_of_words)
mostcommon = wordfreqdist.most_common(30)
print(mostcommon)
```

```python
list_of_bigrams = list(nltk.bigrams(list_of_words))

bigram_freqdist = nltk.FreqDist(list_of_bigrams)

mostcommon = bigram_freqdist.most_common(30)

print(mostcommon)
```

#### Tweets that contain ('aca' or 'affordable care act') OR 'obamacare'
